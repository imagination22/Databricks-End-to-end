# 5. COPY INTO

`COPY INTO` is used for incremental ingestion from cloud storage.

Instead of re-reading all files repeatedly:

* It tracks already processed files.
* Only new files are processed.

This improves:

* Performance
* Cost efficiency

Example conceptually:

```sql
COPY INTO bronze_table
FROM '/data/source/'
FILEFORMAT = CSV
```

---

## Advantages of COPY INTO

### Incremental Loading

Only new files are processed.

### Easier Than Manual Tracking

No need to manually maintain processed file metadata.

### Better for Production

Avoids unnecessary reprocessing.

## 🧱 **RocksDB and State Store**

- **RocksDB** is an **embedded key–value store** developed by Meta (Facebook).  
- In **Spark Structured Streaming**, it’s used **under the hood** by the **state store** to persist streaming state efficiently.  
- The **state store** keeps track of intermediate results for operations like:
  - `groupBy` with aggregations on streaming data  
  - `dropDuplicates` on streams  
  - `mapGroupsWithState` or `flatMapGroupsWithState`  

So yes — **RocksDB *does* store the state store data**.  
It provides:
- Fast read/write performance  
- On-disk persistence (so state survives restarts)  
- Efficient compaction and recovery  

Think of it as the **“memory + disk brain”** of your streaming pipeline — Spark uses it to remember what happened in previous micro‑batches.

---

## ⚙️ **_rescued_data vs _corrupt_record (Deep Clarity)**

Let’s visualize their difference conceptually:

| Aspect | `_rescued_data` | `_corrupt_record` |
|--------|-----------------|-------------------|
| **Purpose** | Captures *extra columns* not defined in the schema. | Captures *invalid rows* that fail parsing. |
| **When It Appears** | When schema evolution or drift occurs (e.g., new fields in JSON). | When data is malformed (e.g., broken CSV line, bad JSON syntax). |
| **Data Type** | JSON string containing unexpected fields. | Raw text of the corrupted row. |
| **Typical Source** | Auto Loader, COPY INTO, or schema inference. | Spark read with `mode="PERMISSIVE"`. |
| **Example** | File has `{id:1, name:"A", country:"IN"}` but schema expects only `id, name`. → `country` goes into `_rescued_data`. | File has `{id:1, name:"A",` (missing closing brace). → Entire line goes into `_corrupt_record`. |

---

### 🧩 **Analogy**
- `_rescued_data` → “Extra luggage tag” — valid data, just unexpected.  
- `_corrupt_record` → “Damaged luggage” — unreadable or broken data.

---

### 🧠 **In Databricks Pipelines**
- **Auto Loader** uses `_rescued_data` to handle schema drift gracefully.  
- **Spark CSV/JSON ingestion** uses `_corrupt_record` to avoid pipeline failure when encountering bad rows.

# Introduction to Auto Loader

Welcome back.

In this lesson, we are going to understand one of the most powerful and modern ingestion features introduced in Databricks:

# Auto Loader

Auto Loader is a very important topic because in real-world enterprise projects, we often deal with:

* millions of files
* continuously arriving data
* schema changes
* cloud storage ingestion

Traditional streaming approaches start becoming inefficient at large scale.

That is exactly the problem Auto Loader solves.

---

# What is Auto Loader?

Auto Loader is:

> A specialized Structured Streaming source designed for efficient, scalable, incremental ingestion of files from cloud storage.

It works with cloud platforms like:

* AWS S3
* Azure Data Lake Storage Gen2 (ADLS Gen2)
* Google Cloud Storage

Its job is simple:

> Detect new files automatically and process them efficiently.

---

# Important Understanding

You may immediately think:

> “But we already used Structured Streaming to process files. Why do we need Auto Loader?”

Excellent question.

Because traditional file streaming has several limitations when the volume becomes very large.

Auto Loader was introduced to overcome those limitations.

---

# Traditional File Streaming Problems

Let us first understand the problems with traditional streaming file sources.

Because unless we understand the problem, we cannot appreciate the solution.

---

# Problem 1 — Continuous Directory Scanning

Traditional streaming continuously scans the entire folder repeatedly.

Example:

Suppose you have:

```text id="2kj0up"
10 million files
```

Every trigger interval:

* Spark scans the entire directory
* checks all files
* identifies new files

This becomes:

* slow
* expensive
* inefficient

Especially in cloud storage.

---

# Why Is It Expensive?

Cloud storage listing operations are not free.

Every directory listing operation costs money and time.

So repeatedly scanning massive folders becomes a serious performance bottleneck.

---

# Problem 2 — In-Memory File Tracking

Traditional streaming keeps track of processed files using:

```text id="s8jvyo"
In-memory map on driver node
```

Meaning:

The Spark driver remembers which files are already processed.

---

# Why Is This Bad?

If you process millions of files:

* memory usage increases
* driver pressure increases
* scalability becomes poor

Eventually:

```text id="e0m3jg"
Driver Memory Issues
```

can happen.

---

# Problem 3 — Manual Schema Definition

Traditional streaming usually requires:

```text id="w4h5o9"
Manual Schema Definition
```

before streaming starts.

Example:

```python id="v1xgk6"
schema = StructType(...)
```

must be written manually.

---

# What Happens If Schema Changes?

Suppose tomorrow incoming JSON contains:

```text id="a7zz9s"
new_column
```

Then problems occur:

* ingestion may fail
* data may be lost
* manual intervention required

This is very common in real-world pipelines.

---

# Auto Loader Solves These Problems

Now let us understand how Auto Loader fixes all these issues.

---

# Auto Loader Feature 1 — File Notification Mode

Instead of scanning entire directories repeatedly:

Auto Loader can use:

# File Notification Mode

This is one of the biggest advantages.

---

# How File Notification Works

Cloud services themselves notify Databricks when new files arrive.

Examples:

| Cloud Platform | Notification Service   |
| -------------- | ---------------------- |
| AWS S3         | S3 Event Notifications |
| Azure          | Event Grid             |

Instead of scanning folders:

```text id="30wh1t"
Cloud Queue → Notification → Databricks
```

---

# Huge Performance Improvement

Because now:

* no full directory scans
* no expensive listing operations
* faster ingestion
* near real-time detection

This makes ingestion highly scalable.

---

# Auto Loader Feature 2 — RocksDB for File Tracking

Traditional streaming used:

```text id="3o5v6k"
In-memory tracking
```

Auto Loader replaces this with:

# RocksDB

---

# What is RocksDB?

RocksDB is:

> A distributed key-value store optimized for large-scale storage.

Instead of storing processed file metadata in driver memory:

Auto Loader stores metadata in RocksDB.

---

# Why Is This Powerful?

Benefits:

* virtually infinite scalability
* no driver memory bottleneck
* reliable deduplication
* scalable file tracking

This is one of the main architectural improvements.

---

# Auto Loader Feature 3 — Automatic Schema Inference

Auto Loader can automatically detect schema from sample files.

This reduces manual effort.

Example:

Instead of manually writing:

```python id="0v5u1q"
StructType(...)
```

Auto Loader can infer schema automatically.

---

# Auto Loader Feature 4 — Schema Evolution

This is extremely important.

In real-world systems:

Schemas change frequently.

Example:

Today:

| id | name |
| -- | ---- |

Tomorrow:

| id | name | city |
| -- | ---- | ---- |

Traditional streaming may fail.

Auto Loader can automatically evolve schema.

---

# What Can Auto Loader Do During Schema Changes?

It supports multiple behaviors:

| Mode        | Behavior                         |
| ----------- | -------------------------------- |
| Add Columns | Automatically evolve schema      |
| Fail        | Stop pipeline                    |
| Rescue Data | Store unexpected data separately |

---

# Rescue Data Column

Auto Loader creates a special column:

```text id="76q7dy"
_rescue_data
```

This column stores:

* invalid records
* unknown columns
* schema mismatch data

instead of failing the pipeline.

This improves resiliency.

---

# Auto Loader Summary

# Auto Loader = Efficient + Scalable + Schema-Aware Ingestion

It provides:

* scalable ingestion
* efficient file tracking
* schema evolution
* cloud notifications
* deduplication

# How to Use Auto Loader

Now let us understand how to implement it.

And the best part is:

# It is extremely simple.

---

# Traditional Streaming Source

Traditional file source looked like this:

```python id="d9g84m"
spark.readStream.format("json")
```

---

# Auto Loader Version

With Auto Loader:

```python id="n9psif"
spark.readStream.format("cloudFiles")
```

That is the major change.

---

# Important Option — File Format

We must specify the actual file format.

Example:

```python id="h13b6c"
.option("cloudFiles.format", "json")
```

This tells Auto Loader:

> “The incoming files are JSON files.”

---

# File Notification Mode Option

If we want notification mode:

```python id="f6d8f7"
.option("cloudFiles.useNotifications", "true")
```

Then Auto Loader uses:

* Event Grid
* S3 Notifications
* cloud queues

instead of directory scanning.

---

# Schema Inference

We can enable schema inference.

Example:

```python id="vxjynw"
.option("cloudFiles.inferColumnTypes", "true")
```

This tells Auto Loader:

> “Try to infer actual data types.”

---

# Important JSON Understanding

JSON is fundamentally:

# Text-Based

So by default:

Auto Loader treats JSON columns as:

```text id="jlwm3r"
STRING
```

unless type inference is enabled.

---

# Initial Auto Loader Output

Without type inference:

| Column            | Type   |
| ----------------- | ------ |
| customer_id       | STRING |
| dob               | STRING |
| created_timestamp | STRING |

Everything becomes string.

---

# _rescue_data Column

Auto Loader also creates:

```text id="6qs1dj"
_rescue_data
```

This is a safety mechanism.

If any row cannot be processed properly:

it gets stored there.

---

# Enabling Infer Column Types

When we enable:

```python id="mf9wqe"
.option("cloudFiles.inferColumnTypes", "true")
```

Auto Loader tries its best to detect types.

Example:

| Column      | Inferred Type |
| ----------- | ------------- |
| customer_id | LONG          |
| others      | STRING        |

---

# Important Limitation

Inference is not always perfect.

Especially if values are quoted.

Example:

```json id="n3hnf6"
"2024-01-01"
```

This looks like string.

So Auto Loader may not identify it as DATE automatically.

---

# Schema Hints

To improve inference, we use:

# Schema Hints

Example:

```python id="cfzklq"
.option("cloudFiles.schemaHints",
"dob DATE, created_timestamp TIMESTAMP")
```

This gives guidance to Auto Loader.

---

# Very Important Understanding

Schema hints are not strict schemas.

They are:

# Suggestions to Auto Loader

This improves accuracy.

---

# Final Schema Result

Now Auto Loader properly detects:

| Column            | Type      |
| ----------------- | --------- |
| dob               | DATE      |
| created_timestamp | TIMESTAMP |

---

# Schema Location

Auto Loader needs a location to store inferred schemas.

Example:

```python id="jjc7xq"
.option("cloudFiles.schemaLocation",
"/path/_schema")
```

This directory stores:

* inferred schema metadata
* schema evolution information

---

# Transformation Layer

After ingestion:

Transformations remain the same.

Because Auto Loader only changes:

# Source Ingestion Layer

The rest of the pipeline usually remains unchanged.

---

# Checkpoint Location

Very important concept.

Streaming queries must maintain separate checkpoint locations.

Example:

```python id="q04jv2"
checkpointLocation = "/checkpoints/customers_autoloader"
```

Never reuse checkpoints between different streams.

Otherwise:

* metadata conflicts
* corrupted state
* unexpected behavior

may happen.

---

# Writing to Delta Table

Example:

```python id="6v4u57"
.writeStream
.format("delta")
```

Auto Loader integrates naturally with Delta Lake.

---

# Incremental File Processing

Suppose:

First file processed:

```text id="i6n5w0"
October_2024.json
```

Then later:

* November file arrives
* December file arrives

Auto Loader only processes:

# NEW FILES

Already processed files are skipped.

Let us understand this carefully.

---

# 8. Understanding the Pattern

Pattern:

```text id="9f3tl0"
customers_2024_*.json
```

This means:

| Part            | Meaning                   |
| --------------- | ------------------------- |
| customers_2024_ | File must begin with this |
| *               | Match any characters      |
| .json           | File must end with .json  |

So Auto Loader will ingest:

✅ `customers_2024_10.json`
✅ `customers_2024_11.json`
✅ `customers_2024_12.json`

But it will ignore:

❌ `customers_2025_01.json`

---

# 9. Wildcard Characters

The lecture also discusses wildcard characters.

These are very important for file filtering.

---

## Asterisk (`*`)

Matches:

* Zero or more characters.

Example:

```text id="v8bj3d"
*.json
```

Matches:

* all JSON files.

---

## Question Mark (`?`)

Matches:

* Exactly one character.

Example:

```text id="if2mml"
file_?.csv
```

Matches:

* `file_1.csv`
* `file_A.csv`

But NOT:

* `file_10.csv`

---

# 10. Complete Auto Loader Example

The ingestion pipeline now becomes:

```python id="s9ws5z"
df = (
    spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "json")
        .option(
            "pathGlobFilter",
            "customers_2024_*.json"
        )
        .load("/input/customers/")
)
```

This ensures:

* Only 2024 JSON files are ingested.

  with
* Previously processed files stored in checkpoints.

New files are identified and processed.

---

# 8. Directory Listing Workflow

```text id="cc8uq5"
Cloud Storage
      ↓
List All Files
      ↓
Compare With Checkpoint
      ↓
Identify New Files
      ↓
Process Incrementally
```

---

# 9. Advantage of Directory Listing Mode

Very easy to use.

No additional setup required.

Works automatically.

---

# 10. Limitation of Directory Listing Mode

Over time:

* Millions of files may accumulate.

Then every execution requires:

* Large directory scans.

This reduces efficiency.

Especially for:

* Massive production systems.

---

# 11. File Notification Mode

Databricks introduced a much more efficient approach:

# File Notification Mode

This is the recommended approach today.

---

# 12. How File Notification Mode Works

Instead of scanning folders repeatedly:

Cloud storage itself generates notifications whenever:

* New files arrive.

---

# 13. Architecture of File Notification Mode

```text id="o6is4t"
New File Arrives
        ↓
Cloud Storage Generates Event
        ↓
Event Broker / Queue
        ↓
Auto Loader Reads Queue
        ↓
Process New Files
```

---

# 14. Why File Notification Mode Is Faster

Because:

* Auto Loader no longer scans entire directories.
* It simply checks queue notifications.

Only new files are tracked.

This makes ingestion:

* Highly scalable
* Highly efficient

---

# 15. Trade-Off of File Notification Mode

Although efficient:

* Additional setup is required.

Examples:

* Queue services
* Event subscriptions
* Permissions
* Notification infrastructure

But for enterprise systems:

* This is usually worth it.

# 40. Complete Production Auto Loader Template

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType()),
    StructField("amount", DoubleType())
])

df = (
    spark.readStream
         .format("cloudFiles")
         .schema(schema)
         .option("cloudFiles.format", "json")
         .option(
             "cloudFiles.schemaLocation",
             "/schema/sales"
         )
         .option(
             "cloudFiles.schemaEvolutionMode",
             "rescue"
         )
         .load("/input/sales")
)

query = (
    df.writeStream
      .format("delta")
      .option(
          "checkpointLocation",
          "/chk/sales"
      )
      .option("mergeSchema", "true")
      .trigger(availableNow=True)
      .toTable("bronze.sales")
)

query.awaitTermination()
```
