# Chapter 1 — Understanding PII Data

Now before learning how to protect PII data, first we must clearly understand:

> What exactly is PII data?

---

# What is PII?

PII stands for:

## Personally Identifiable Information

This means:

> Any data that can directly or indirectly identify a specific individual.

This definition is extremely important.

The key point here is:

If some data can help us figure out which person the data belongs to, then it is considered PII data.

---

# Simple Way to Remember PII

Think about this question:

> “Can this data help identify a person?”

If the answer is YES → it is PII.

---

# Two Main Types of PII

PII is mainly divided into two categories:

1. Direct PII
2. Indirect PII

Let us understand both carefully.

---

# 1. Direct PII

Direct PII means:

> A single piece of data alone is enough to identify a person.

So even if we have only one column or one value, we can identify the individual.

---

# Examples of Direct PII

Examples include:

* Full Name
* Email Address
* Phone Number
* Social Security Number (SSN)
* Passport Number

---

# Why is it called “Direct” PII?

Because:

> The data uniquely identifies a person on its own.

For example:

If I give you a passport number, you can directly trace it back to one specific person.

No additional information is required.

That is why it is called Direct PII.

---

# Memory Trick — Direct PII

## “One Value = One Person”

If a single value identifies one individual directly → Direct PII.

---

# 2. Indirect PII

Now let us understand indirect PII.

Indirect PII means:

> The data alone cannot identify a person, but when combined with other information, it can identify someone.

This is very important.

Many people think only names and phone numbers are sensitive.

But indirect information can also become dangerous when combined together.

---

# Examples of Indirect PII

Examples include:

* Date of Birth
* IP Address
* Device ID
* ZIP Code
* Location Data

---

# Understanding with an Example

Suppose we only have:

* Date of Birth

Using only DOB, we usually cannot identify a person.

Now suppose we combine:

* Date of Birth
* IP Address
* Device Information

Together, these may point to a specific individual.

So individually the data may look harmless.

But combined together, it becomes sensitive.

That is why indirect PII is also very important.

---

# Memory Trick — Indirect PII

## “One Value = Not Enough

But Combined = Dangerous”

---

# Direct vs Indirect PII

| Type         | Meaning                           | Example          |
| ------------ | --------------------------------- | ---------------- |
| Direct PII   | Directly identifies a person      | Passport Number  |
| Indirect PII | Needs combination with other data | DOB + IP Address |

---

# Why Handling PII Properly is Important

Now the next big question is:

> Why do we need special techniques to handle PII data?

Why not store it like normal data?

The answer is because PII data is highly sensitive.

And mishandling it can create massive problems.

There are mainly three major reasons.

---

If attackers steal PII data and understand it properly, they can misuse it for:

* Identity theft
* Fraud
* Financial crimes
* Phishing attacks
* Illegal activities

And because of this:

> Organizations can lose millions or even billions of dollars.

The company may need to:

* Pay penalties
* Handle lawsuits
* Compensate users
* Recover systems
* Investigate incidents

So data breaches are extremely expensive.

---

# Real Industry Understanding

That is why companies invest heavily in:

* Security
* Encryption
* Monitoring
* Compliance
* Access control

---

# 2. Regulatory Requirements and Compliance

The second major reason is:

## Legal compliance.

Organizations that store PII data are required to follow strict regulations.

These regulations define:

* How data should be collected
* How it should be stored
* Who can access it
* How long it should be retained
* How it should be deleted

---

Used mainly in the United States for healthcare-related information.

## GDPR

Used heavily in Europe for personal data protection.

There are many other regulations as well depending on country and industry.

---

# Important Concept

If organizations fail to comply with regulations:

* Heavy fines may occur
* Legal action may happen
* Business operations may be affected

So compliance is not optional.

It is mandatory.

---

# 3. Reputational Damage

The third reason is:

## Loss of trust.

If customer personal data gets leaked:

* Customers lose confidence
* Brand reputation gets damaged
* Users may stop using the platform

And rebuilding trust is extremely difficult.

Sometimes the reputation damage is even more harmful than the financial loss.

---

# Important Industry Mindset

In cybersecurity and data engineering:

> Protecting customer trust is as important as protecting systems.

# Chapter 2 — Minimize, Classify, Protect, Control, Monitor, and Delete PII Data

# Introduction

Alright everyone, now that we understand what PII data is and why it is important, the next question becomes:

> “How do organizations actually handle PII data properly?”

Because understanding PII is only the first step.

The real challenge is:

* How to store it
* How to secure it
* How to control access
* How to monitor usage
* And how to safely delete it when required

So in this section, we are going to understand the complete lifecycle of handling PII data.

---

# High-Level Lifecycle of Handling PII Data

There are mainly six major stages:

1. Minimize
2. Classify
3. Protect
4. Control
5. Monitor
6. Delete

Think of this as a pipeline or framework for handling sensitive data.

---

# Big Picture Mental Model

## “PII Data Security Lifecycle”

```text
Minimize → Classify → Protect → Control → Monitor → Delete
```

This is the complete journey of securely handling PII data.

---

# 1. Minimize — The Best Security is Not Storing Data

Now the first principle is extremely important.

And honestly, this is one of the smartest security techniques.

## If you do not need PII data,

## then do not ingest it.

This sounds simple, but it is a very powerful concept.

---

# Why Minimization Matters

The more PII data we store:

* The higher the risk
* The higher the compliance burden
* The higher the storage responsibility
* The higher the attack surface

So organizations should always ask:

> “Do we really need this data?”

---

# Example Scenario

Suppose we are building an analytics platform.

We receive customer data containing:

| Column       |
| ------------ |
| Customer ID  |
| Country      |
| Email        |
| Phone Number |

Now suppose for analytics purposes, we only need:

* Customer ID
* Country

We do NOT need:

* Email
* Phone Number

Then what should we do?

Simple.

We exclude those columns during ingestion itself.

---

# Important Principle

## “Unused PII is unnecessary risk.”

If we never ingest sensitive data:

* We do not need to protect it
* We do not need to monitor it
* We do not need to delete it later

This greatly simplifies the architecture.

---

# Two Levels of Minimization

There are two ways to minimize PII.

---

# Level 1 — Exclude Entire Tables

Suppose some tables contain only sensitive information and are not needed.

Then:

> Do not ingest those tables into the data lake.

Very straightforward.

---

# Level 2 — Exclude Specific Columns

Sometimes the table itself is useful.

But only some columns are sensitive.

For example:

| Needed      | Not Needed   |
| ----------- | ------------ |
| Customer ID | Email        |
| Country     | Phone Number |

Then:

> Ingest only the required columns.

This is a very common enterprise practice.

---

# Memory Trick — Minimize

# “If you don’t need it, don’t store it.”

This single sentence is enough to remember the entire minimization concept.

---

# 2. Classify — Identify Sensitive Data Properly

Now suppose we ARE storing PII data.

Then the next important step is:

# Classification

This means:

> Clearly identifying which columns are sensitive.

Because if we do not know which data is PII:

* We cannot secure it properly
* We cannot audit it properly
* We cannot apply policies properly

---

# Why Classification is Important

Imagine a database with 500 columns.

How will security systems know:

* Which column contains email?
* Which column contains phone number?
* Which column contains medical data?

That is why classification is essential.

---

# What Happens During Classification?

We identify:

* Which columns are PII
* Which columns are PHI
* Which columns are non-sensitive

And then we apply metadata or tags.

---

# Example of Classification

| Column         | Classification |
| -------------- | -------------- |
| customer_id    | Non-PII        |
| email          | PII            |
| phone_number   | PII            |
| medical_report | PHI            |

---

# Tagging Columns

After classification, we usually attach tags.

For example:

| Column         | Tag      |
| -------------- | -------- |
| email          | PII=True |
| phone_number   | PII=True |
| medical_report | PHI=True |

---

# Important Concept — Metadata Driven Security

Modern data platforms use metadata heavily.

That means:

> Security systems rely on tags and classifications.

Using tags, systems can:

* Automatically mask data
* Apply encryption
* Restrict access
* Trigger monitoring alerts

---

# Custom Classification Tags

Organizations can create custom tags based on business needs.

Examples:

| Tag        | Meaning                   |
| ---------- | ------------------------- |
| Sensitive  | Confidential data         |
| Financial  | Banking data              |
| Health     | Medical data              |
| Restricted | High-security information |

So classification systems are customizable.

---

# Automatic PII Detection

Now modern platforms like Databricks provide intelligent features.

If enabled at the catalog level:

* Databricks can automatically scan columns
* Detect possible PII
* Add tags automatically

This is extremely useful for large-scale enterprise systems.

---

# Important Industry Insight

In real-world organizations:

> Manual classification does not scale.

That is why automated data discovery and classification tools are becoming very important.

---

# Memory Trick — Classification

# “You cannot protect what you cannot identify.”

This is one of the most important cybersecurity principles.

---

Now after identifying PII data, the next step is:

# 4. Control — Restrict Access Properly

Now even after securing the data, one more important question remains:

> “Who should be allowed to see the data?”

This is where access control comes into the picture.

---

# Core Idea of Access Control

Not everyone should access everything.

Different users should see different data depending on their role.

---

# Examples

| User          | Access       |
| ------------- | ------------ |
| Data Analyst  | Masked Email |
| Security Team | Full Email   |
| Finance Team  | No Access    |

---

# Techniques Used

Two very common techniques are:

---

# Column-Level Masking

Specific columns are hidden or partially masked.

Example:

```text
john@gmail.com
```

May appear as:

```text
jo****@gmail.com
```

---

# Row-Level Filtering

Different users can see different rows.

For example:

* India team sees India customers
* Europe team sees Europe customers

---

# Role-Based Access Control (RBAC)

Access is controlled based on roles.

Example roles:

* Admin
* Analyst
* Auditor
* Security Engineer

This is a standard enterprise approach.

---

# Memory Trick — Control

# “Not everyone should see everything.”

---

# 5. Monitor — Track Usage and Access

Now after storing and controlling data:

We must continuously monitor its usage.

Because security is not a one-time setup.

---

# Why Monitoring is Important

Organizations must know:

* Who accessed data
* When it was accessed
* Which table was queried
* Which user downloaded data

This helps detect:

* Suspicious activity
* Insider threats
* Data misuse
* Compliance violations

---

# Audit Logs

Platforms like Databricks maintain audit logs.

These logs track operations related to:

* Data access
* Queries
* Modifications
* Permissions

---

# Monitoring Dashboards

Organizations often build dashboards for:

* Access tracking
* Compliance reporting
* Security alerts
* User activity monitoring

---

# Important Security Principle

# “If you cannot monitor it, you cannot secure it.”

---

# 6. Delete — Support Proper Data Deletion

Now the final stage is:

# Deletion

In many analytics projects:

Data is rarely deleted.

But certain regulations require:

> Users must be able to request deletion of their personal data.

---

# Example Scenario

Suppose a user deletes their account.

Then the organization may also need to:

* Remove personal data
* Delete historical records
* Clean downstream systems

And this often must happen within a fixed legal time window.

---

# Important Challenge

Deletion becomes difficult because data may exist in:

* Data lakes
* Warehouses
* Backups
* Logs
* Analytics tables

That is why deletion pipelines must be carefully designed.

---

# Engineering Requirement

Good data systems should support:

* User-level deletion
* Traceability
* Cleanup automation
* Compliance reporting

---

# Memory Trick — Delete

# “Users should have the right to disappear.”

---

# Important Understanding

These three techniques solve different problems.

Not every project uses the same approach.

The correct choice depends on:

* Business requirements
* Compliance rules
* Analytics needs
* Security requirements
* Whether re-identification is needed

---

# 1. Pseudonymization

Let us start with:

# Pseudonymization

---

# What is Pseudonymization?

In pseudonymization:

> We replace original sensitive values with substitute values or identifiers.

The original data is hidden.

But the data still remains useful for processing or analytics.

---

# Key Idea

Instead of storing:

| Original Value                          |
| --------------------------------------- |
| John Smith                              |
| [john@gmail.com](mailto:john@gmail.com) |

We replace it with:

| Protected Value |
| --------------- |
| X82KD9          |
| A91PQ7          |

---

# Important Characteristic

Pseudonymization usually works:

# Row by row

Meaning:

Each sensitive value gets transformed individually.

---

# Re-Identification in Pseudonymization

Depending on the technique:

* Re-identification may be possible
* Or may not be possible

This depends on which pseudonymization method is used.

---

# Two Main Techniques Inside Pseudonymization

1. Hashing
2. Tokenization

---

# Big Picture

```text id="0rr44u"
Pseudonymization
      ├── Hashing
      └── Tokenization
```

---

# 2. Anonymization

Now let us understand anonymization.

---

# What is Anonymization?

In anonymization:

> We completely remove the ability to identify individuals.

Instead of carrying detailed PII data forward,

we only keep:

* Aggregated data
* Categories
* Summaries
* Statistical information

---

# Example

Instead of storing:

| Name | Age |
| ---- | --- |
| John | 32  |
| Mike | 34  |

We may only store:

| Age Group | Count |
| --------- | ----- |
| 30–35     | 2     |

---

# Important Difference

Pseudonymization:

* Protects rows individually

Anonymization:

* Protects the entire dataset

---

# Most Important Property

# Re-identification is NOT possible.

Once data is anonymized properly:

> Original identity cannot be reconstructed.

---

# Why Anonymization is Powerful

This is extremely useful for:

* Analytics
* Reporting
* Research
* Trend analysis

Where individual identity is not required.

---

# 3. Encryption

Now the third technique is:

# Encryption

---

# What is Encryption?

Encryption means:

> We convert readable data into unreadable encrypted form using cryptographic keys.

---

# Example

Original:

```text id="8vvjlwm"
john@gmail.com
```

Encrypted:

```text id="fpp4fz"
X7A91L0M2Q...
```

---

# How Encryption Works

There are two main steps:

1. Encrypt while storing
2. Decrypt while retrieving

---

# Important Characteristic

Encryption is designed for:

# Reversible protection

Meaning:

If we have the correct key,

we can recover the original data.

---

# Re-Identification in Encryption

Unlike anonymization:

# Re-identification is definitely possible.

That is actually the purpose of encryption.

---

# High-Level Comparison

| Technique | Reversible? | Row Level? | Analytics Friendly? |
|---|---|---|
| Pseudonymization | Sometimes | Yes | Yes |
| Anonymization | No | No | Very High |
| Encryption | Yes | Yes | Limited |

---

# Important Mental Model

---

# Pseudonymization

## “Hide the identity.”

---

# Anonymization

## “Remove the identity.”

---

# Encryption

## “Lock the identity.”

---

# Deep Dive into Pseudonymization

Now let us go deeper into pseudonymization techniques.

---

# Technique 1 — Hashing

This is one of the most common techniques.

---

# What is Hashing?

Hashing means:

> Converting data into a fixed-length hash value using mathematical algorithms.

---

# Example

Original:

```text id="gw1rbg"
john@gmail.com
```

Hashed:

```text id="4wzv6h"
5f4dcc3b5aa765d61d8327deb882cf99
```

---

# Important Property of Hashing

Hashing is:

# One-way

Meaning:

* Original → Hash is possible
* Hash → Original is NOT possible

---

# Popular Hashing Algorithms

Common examples include:

* MD5
* SHA1
* SHA256
* SHA512

---

# Important Note

Different algorithms generate:

* Different lengths
* Different security levels
* Different performance characteristics

---

# Hashing Demo — Understanding the Flow

Now let us understand the practical implementation shown in the demo.

---

# Reading the Source Table

The instructor first reads the original customer table.

This table contains PII columns like:

| Column    |
| --------- |
| full_name |
| email     |
| phone     |
| ssn       |

---

# Displaying Sample Rows

Before hashing:

The original readable values are displayed.

This helps us compare:

* Before hashing
* After hashing

---

# Applying Hash Functions

Now different hash functions are applied to different columns.

---

# Example — SHA2

For full_name:

```python id="4tp5c5"
sha2(col("full_name"), 256)
```

---

# Understanding the Parameters

Two inputs are provided:

1. Column name
2. Bit length

---

# Important Correction

The instructor mentioned “bytes” verbally,

but technically:

# SHA-256 means 256 bits.

Not bytes.

---

# Example — MD5

For email:

```python id="9e4rj3"
md5(col("email"))
```

---

# Example — SHA1

For phone number:

```python id="x0mx0s"
sha1(col("phone"))
```

---

# Example — SHA512

For SSN:

```python id="k93mxk"
sha2(col("ssn"), 512)
```

This creates a much longer hash output.

---

# Visual Observation

Different algorithms create:

* Different output lengths
* Different hash patterns

For example:

| Algorithm | Length    |
| --------- | --------- |
| MD5       | Short     |
| SHA1      | Medium    |
| SHA256    | Longer    |
| SHA512    | Very Long |

---

# Important Concept — Hash Collision

Sometimes theoretically:

Two different values may generate the same hash.

This is called:

# Collision

---

# Real-World Understanding

Strong algorithms like SHA256 make collisions extremely rare.

For most practical enterprise workloads:

They are not a major issue.

---

# MD5 vs SHA

The instructor mentions an important tradeoff.

---

# MD5

Advantages:

* Fast
* Widely used

Disadvantages:

* Less secure
* Older algorithm

---

# SHA Family

Advantages:

* Stronger security
* Better collision resistance

Disadvantages:

* Slightly slower

---

# Important Engineering Tradeoff

In real projects:

We balance:

* Performance
* Security
* Storage
* Scalability

---

# What Happens After Hashing?

The original values disappear.

Example:

| Original   | Hashed          |
| ---------- | --------------- |
| John Smith | 8f14e45fceea... |

---

# Most Important Property

From the hash value:

# Original data cannot be reconstructed.

This is because hashing loses original information permanently.

---

# Important Clarification

Hashing is NOT encryption.

Because:

Encryption is reversible.

Hashing is not.

---

# Why Hashing is Useful

Even though original values disappear,

we can still perform certain operations.

---

# Example 1 — Matching Records

Suppose two tables hash emails using the same algorithm.

Then:

* Same email
* Produces same hash

So tables can still be joined.

---

# Example 2 — Distinct Count

We can still calculate:

* Unique users
* Frequency analysis
* Duplicate detection

Without exposing actual PII.

---

# This is Extremely Important

Hashing allows:

# Analytics without exposing identity.

---

# Final Result

The hashed table is written into the target system.

Now the stored values are protected hash values instead of raw PII.

---

# Memory Trick — Hashing

# “Same Input → Same Hash

But Hash → Original Impossible”

---

# Important Limitation of Hashing

Hashing is good when:

* We do NOT need original values later

But if business requires recovery of original data:

Hashing alone is insufficient.

That is where tokenization comes into the picture.

---

# Technique 2 — Tokenization

Now we move to the second pseudonymization technique:

# Tokenization

---

# What is Tokenization?

In tokenization:

> Original sensitive data is replaced with tokens or IDs.

But unlike hashing:

# Original values are stored separately.

---

# Example

Original:

| Customer   |
| ---------- |
| John Smith |

Tokenized:

| Token     |
| --------- |
| CUST_1001 |

---

# Separate Lookup Table

A secure lookup table stores:

| Token     | Original Value |
| --------- | -------------- |
| CUST_1001 | John Smith     |

---

# Very Important Difference

Hashing:

* Original value lost

Tokenization:

* Original value preserved separately

---

# How Access is Controlled

Normal users:

* See only tokens

Privileged users:

* Can access lookup table
* Can reconstruct original data

---

# Important Security Principle

# “Separate identity from usage.”

---

# Why Tokenization is Powerful

This technique is heavily used in:

* Banking systems
* Payment systems
* Healthcare systems
* Enterprise analytics

Because:

Business processes can continue using tokens without exposing raw data.

---

# Real-World Example — Credit Cards

Instead of storing:

```text id="j8j7sj"
4532-XXXX-XXXX-1234
```

Systems may store:

```text id="xwbr0v"
TOKEN_847291
```

The real card number exists only in a highly secured vault.

---

# Hashing vs Tokenization

| Feature                  | Hashing | Tokenization |
| ------------------------ | ------- | ------------ |
| Reversible               | No      | Yes          |
| Lookup Table Needed      | No      | Yes          |
| Same Input → Same Output | Yes     | Usually Yes  |
| Original Recoverable     | No      | Yes          |

---

# Golden Concepts to Remember

---

# Pseudonymization

# “Replace identity while preserving usability.”

---

# Hashing

# “One-way mathematical masking.”

---

# Tokenization

# “Replace with IDs and secure the mapping.”

---

# Anonymization

# “Remove identity permanently.”

---

# Encryption

# “Lock data with keys.”

---

# Final Mental Model

```text id="73p9vg"
Protecting PII
     ├── Hide Identity → Pseudonymization
     │        ├── Hashing
     │        └── Tokenization
     │
     ├── Remove Identity → Anonymization
     │
     └── Lock Identity → Encryption
```

# Introduction

# Anonymization

# How to completely remove identity from the data.

* Trends
* Categories
* Statistics
* Summaries
* Group-level insights

And that is exactly where anonymization becomes very powerful.

Anonymization can broadly be divided into two categories:

1. Data Suppression

---

# Big Picture Mental Model

```text id="ymjlwm"
Anonymization
     ├── Data Suppression
     └── Generalization
              ├── Categorical Generalization
              ├── Binning

```

---

# Important Goal of Anonymization

The core purpose is:

# “Keep analytics useful while removing identity.”

---

# Most Important Property

Unlike encryption or tokenization:

# Proper anonymization is irreversible.

Once data is anonymized correctly:

> We cannot recover the original identity.

---

# 1. Data Suppression

Let us first understand:

# Data Suppression

---

# What is Data Suppression?

Data suppression means:

> Hiding or excluding sensitive columns from users.

Instead of giving access to the original table,

we create:

# A restricted view

which contains only non-sensitive columns.

---

# Simple Example

Suppose original table contains:

| customer_id | name | email | country |
| ----------- | ---- | ----- | ------- |

Now for normal analysts,

we create a view containing only:

| customer_id | country |
| ----------- | ------- |

---

# Important Understanding

The original data still exists.

But:

> Unauthorized users cannot see it.

---

# How Suppression Works

The flow is:

```text id="ygp4gq"
Original Table
       ↓
Create Restricted View
       ↓
Provide View Access
       ↓
Hide PII Columns
```

---

# Important Difference from Minimization

This is very important.

Students often confuse:

* Minimization
* Suppression

---

# Minimization

During ingestion itself:

* PII columns are NOT brought into target.

---

# Suppression

PII data EXISTS in target table.

But:

* Access is restricted through views.

---

# Memory Trick

# “Minimization removes data.

# Suppression hides data.”

---

# Demo — Data Suppression

Now let us understand the practical demo.

---

# Step 1 — Existing Table Already Contains PII

The instructor uses the classified customer table created earlier.

This table still contains:

* name
* email
* phone
* SSN

All PII exists inside the table.

---

# Step 2 — Create a View

A new view is created selecting only:

```python id="mzjbrf"
customer_id
country
```

---

# Important Concept

Views behave like virtual tables.

They do not duplicate the data.

Instead:

> They provide controlled visibility.

---

# Step 3 — Grant Access to View

Now:

* Normal users get access to the view
* Admin users get access to original table

---

# Result

Unauthorized users:

* Cannot see PII

Authorized users:

* Can access complete data

---

# Why Suppression is Useful

This is heavily used in enterprises because:

* Same dataset can serve multiple teams
* Access becomes role-based
* PII exposure reduces significantly

---

# Important Security Principle

# “Users should only see what they truly need.”

---

# 2. Generalization

Now we move to the second anonymization category:

# Generalization

---

# What is Generalization?

Generalization means:

> Replacing precise information with broader categories.

This reduces precision,

making identification difficult.

But analytics still remains useful.

---

# Core Idea

Move from:

# Specific → General

---

# Why Generalization Works

Exact values help identify individuals.

General categories usually do not.

So by reducing precision:

> We reduce identifiability.

---

# Types of Generalization

There are four techniques discussed:

1. Categorical Generalization
2. Binning
3. IP Address Truncation
4. Rounding

---

# 2.1 Categorical Generalization

Let us start with:

# Categorical Generalization

---

# What is It?

Here:

> Specific values are replaced with broader categories.

---

# Example

Original:

| Job Title         |
| ----------------- |
| Software Engineer |
| HR Manager        |
| Product Manager   |

Generalized:

| Category    |
| ----------- |
| Engineering |
| Management  |

---

# Important Observation

We lose exact job identity.

But analytics still remains meaningful.

---

# Demo — Categorical Generalization

The instructor reads the employees table.

Focus column:

# job_title

---

# Mapping Logic

Specific job titles are mapped to generalized categories.

Example:

| Original          | Generalized |
| ----------------- | ----------- |
| Software Engineer | Engineering |
| Product Manager   | Management  |
| HR Manager        | Management  |

---

# Important Understanding

Now the data no longer reveals:

* Exact role
* Exact designation

But still supports analytics like:

* Number of engineering employees
* Number of management employees

---

# Result

The final table contains:

| job_title   |
| ----------- |
| Engineering |
| Management  |

instead of detailed roles.

---

# Memory Trick

# “Reduce specificity, preserve usefulness.”

---

# 2.2 Binning

Now the next technique is:

# Binning

---

# What is Binning?

Binning means:

> Replacing exact values with ranges or groups.

---

# Most Common Example

Date of birth → Age groups

---

# Example

Original DOB:

```text id="0v2wd4"
1995-07-10
```

Instead of storing DOB:

We calculate:

```text id="jz05xr"
Age = 29
```

Then convert into:

```text id="77bhgx"
25-35
```

---

# Why This Helps

Now we know:

* Approximate age range

But NOT:

* Exact birthday

---

# Demo — Binning

The instructor reads the customers table.

Focus column:

# date_of_birth

---

# Step 1 — Calculate Age

Age is calculated using:

* Current date
* DOB

---

# Step 2 — Create Age Groups

Example logic:

| Age   | Group    |
| ----- | -------- |
| <18   | Under 18 |
| 18–25 | 18–25    |
| 26–40 | 26–40    |

---

# Step 3 — Remove Exact Age

The exact DOB and exact age are discarded.

Only age_range remains.

---

# Final Result

The final dataset contains:

| age_range |
| --------- |
| 18–25     |
| 26–40     |

instead of exact DOB.

---

# Analytics Still Works

We can still analyze:

* Customer distribution by age group
* Marketing demographics
* Trend analysis

Without exposing identity.

---

# Important Privacy Principle

# “Analytics often needs patterns, not identity.”

---

# 2.3 IP Address Truncation

Now let us understand a very specific technique:

# Truncating IP Addresses

---

# Why IP Address is Sensitive

IP addresses are considered:

# Indirect PII

Because they can help identify:

* Devices
* Users
* Locations

---

# Example IP

After:

```text id="7p2jpy"
192.168.1.0
```

Rounded Salary:

```text id="xvsj1c"

```

---

# Why This Helps

Now we understand:

* Approximate salary range

But not:

* Exact compensation

---

# Demo — Salary Rounding

The instructor reads the employees table.

Focus column:

# salary

---

# Transformation

The last digits are replaced with zeros.

---

# Final Result

The dataset now contains approximate salary values.

---

# Analytics Still Works

We can still calculate:

* Salary ranges
* High-income vs low-income groups
* Compensation distribution

Without revealing exact salaries.

---

# Important Tradeoff

Rounding slightly reduces accuracy.

But greatly improves privacy.

---

# Memory Trick

# “Preserve trends, hide precision.”

---

# Full Generalization Summary

| Technique                  | Purpose                  |
| -------------------------- | ------------------------ |
| Categorical Generalization | Specific → Category      |
| Binning                    | Exact values → Ranges    |
| IP Truncation              | Hide exact device        |
| Rounding                   | Reduce numeric precision |

---

# Most Important Understanding

Notice something very important.

In anonymization:

> We intentionally sacrifice detail to improve privacy.

This is one of the biggest tradeoffs in data engineering.

---

# Privacy vs Utility Tradeoff

More detail:

* Better analytics
* Lower privacy

Less detail:

* Stronger privacy
* Reduced accuracy

Good data engineers balance both.

---

# Golden Rules to Remember

---

# Suppression

# “Hide sensitive columns.”

---

# Generalization

# “Reduce precision to reduce identification.”

---

# Binning

# “Convert exact values into ranges.”

---

# IP Truncation

# “Hide exact devices.”

---

# Rounding

# “Keep approximate value, remove precision.”

---

# Important Final Insight

Anonymization is extremely valuable because:

* Analytics still works
* Compliance becomes easier
* Privacy risk reduces drastically

And in many business cases:

> Aggregated information is enough.

---

# Final Mental Model

```text id="94c8o7"
Anonymization
      ↓
Remove Precision
      ↓
Reduce Identifiability
      ↓
Preserve Analytics
```

# Chapter 6 — Encryption for PII Data Protection

# Introduction

# Encryption

Until now we understood:

* Minimization
* Classification
* Pseudonymization
* Anonymization

Now we are going to learn:

# How to lock data cryptographically.

This is one of the most important techniques used in:

* Enterprise security
* Banking systems
* Healthcare systems
* Government systems
* Cloud platforms

Because even if attackers steal encrypted data:

> They still should not be able to understand it.

---

# What is Encryption?

Encryption means:

> Converting readable data into unreadable encrypted form using cryptographic algorithms and keys.

---

# Core Idea

Original Data:

```text id="1b0gbf"
john@gmail.com
```

Encrypted Data:

```text id="z06d9l"
A8F72J9XQW...
```

---

# Important Concept

Encrypted data looks meaningless.

But:

If we have:

* Correct encryption algorithm
* Correct encryption key

Then:

# We can decrypt the data back.

---

# Important Difference from Hashing

Students often confuse:

* Hashing
* Encryption

---

# Hashing

* One-way
* Not reversible

---

# Encryption

* Two-way
* Reversible with key

---

# Memory Trick

# “Hashing hides forever.

# Encryption locks temporarily.”

---

# Three Major Types of Encryption in Data Engineering

When handling PII data,

there are mainly three important encryption areas:

1. Column Encryption
2. Encryption at Rest
3. Encryption in Motion

---

# Big Picture Mental Model

```text id="a0x4mb"
Encryption
     ├── Column Encryption
     ├── Encryption at Rest
     └── Encryption in Motion
```

---

# 1. Column Encryption

Let us start with:

# Column Encryption

---

# What is Column Encryption?

Here:

> Specific PII columns are encrypted before storing into the table.

---

# Example

Instead of storing:

| full_name  |
| ---------- |
| John Smith |

We store:

| full_name   |
| ----------- |
| X92KLS82... |

---

# Important Observation

Only selected sensitive columns are encrypted.

Non-PII columns may remain normal.

---

# Core Flow

```text id="c0ivk2"
Original Data
      ↓
Encrypt Sensitive Columns
      ↓
Store Encrypted Values
      ↓
Decrypt Only When Needed
```

---

# Reversibility

Unlike hashing:

# Column encryption is reversible.

If we use:

* Correct key
* Correct algorithm

Then original data can be recovered.

---

# Important Real-World Usage

Column encryption is heavily used for:

* SSN
* Credit cards
* Emails
* Phone numbers
* Medical records

# Demo — Reading Source Table

# Step 1 — Define Encryption Key

Inside Databricks, under the **Permissions Section**, we can assign permissions to groups.

Example:

## Admin Group

```text
Privileges:
- SELECT
- INSERT
- UPDATE
- DELETE
- MODIFY
```

Admins can do everything.

---

## Analyst Group

```text
Privileges:
- SELECT only
```

Analysts can only read the data.

They cannot:

* modify data
* delete records
* update tables

---

# Important Understanding

RBAC works at multiple levels:

| Level         | Can Control? |
| ------------- | ------------ |
| Catalog Level | Yes          |
| Schema Level  | Yes          |
| Table Level   | Yes          |

So RBAC provides **broad-level control**.

But what if we want more granular control?

For example:

* Show only certain rows
* Hide only specific columns

That is where:

* Row Filters
* Column Masking

come into picture.

---

# 2. Column Masking

Now let us understand one of the most important PII protection techniques:

# Column Masking

Column masking means:

> “Show partially hidden data depending on the user.”

Example:

Instead of showing:

```text
Shubham
```

we show:

```text
S***
```

Or instead of:

```text
9876543210
```

we show:

```text
******3210
```

The real data exists in the table.

But users see only masked values.

---

# How Column Masking Works

Suppose we have a customers table with original data.

Example:

| full_name | email                                         | phone      |
| --------- | --------------------------------------------- | ---------- |
| Shubham   | [shubham@gmail.com](mailto:shubham@gmail.com) | 9876543210 |

Now we create a masking function.

---

# Full Name Masking Function

Example logic:

```sql
IF user belongs to admin group
THEN show original value
ELSE show first character + ***
```

So:

| User Type | Output  |
| --------- | ------- |
| Admin     | Shubham |
| Non-admin | S***    |

---

# Understanding the Logic

The masking function checks:

```sql
is_account_group_member('admin')
```

Meaning:

> “Is the current logged-in user part of the admin group?”

If YES:

* return original value

If NO:

* return masked value

---

# Important Realization

Masking depends on:

* Current user
* User group
* Defined policy

So the same table can show:

* real values to one user
* masked values to another user

---

# Different Types of Masking

Different columns use different masking strategies.

---

# 1. Name Masking

```text
Shubham → S***
```

Keep first letter only.

---

# 2. Email Masking

```text
shubham@gmail.com
→ s***@gmail.com
```

Keep domain visible.

Because:

* domain is usually not PII
* useful for analytics

---

# 3. Phone Number Masking

```text
9876543210
→ ******3210
```

Only last 4 digits visible.

Common industry practice.

---

# 4. SSN Masking

Partial visibility.

Example:

```text
XXX-XX-1234
```

---

# 5. Date of Birth Masking

Instead of full DOB:

```text
1995-10-21
```

Show only:

```text
1995
```

---

# 6. Address Masking

Completely hidden.

```text
********
```

---

# 7. City Masking

```text
Mumbai → Mu***
```

Only partial visibility.

---

# Very Important Concept

Creating a masking function alone does NOT apply masking.

First:

```text
Function is created
```

Then:

```text
Function is attached to a column
```

using:

```sql
ALTER TABLE
ALTER COLUMN
SET MASK
```

This creates the connection between:

* table
* column
* masking function

---

# Final Result

Different users see different outputs.

Example:

| User    | What They See |
| ------- | ------------- |
| Admin   | Original data |
| Analyst | Masked data   |

---

# Key Concept to Remember

# Column Masking = RBAC + Data Transformation

Because:

* RBAC decides WHO the user is
* Masking decides WHAT they see

# 3. Row Filtering

Now let us move to another powerful technique:

# Row Filtering

Column masking hides values.

But row filtering hides entire rows.

---

# What is Row Filtering?

It means:

> “Show only specific rows depending on the user.”

Example:

Suppose table contains data from:

* India
* USA
* UK
* Canada

Maybe Indian employees should only see India data.

That is row filtering.

---

# How Row Filtering Works

We create a function that returns:

```text
TRUE or FALSE
```

* TRUE → row visible
* FALSE → row hidden

---

# Example Logic

Suppose:

Managers can see everything.

Others can only see India rows.

Logic:

```sql
IF user belongs to manager group
THEN TRUE

ELSE IF country = 'India'
THEN TRUE

ELSE FALSE
```

---

# Understanding the Flow

The filter runs row by row.

For each row:

| Country | Result |
| ------- | ------ |
| India   | TRUE   |
| USA     | FALSE  |
| UK      | FALSE  |

So non-manager users see only India rows.

---

# Applying Row Filter

Unlike column masking:

```sql
SET MASK
```

Row filtering uses:

```sql
SET ROW FILTER
```

on a specific column.

---

# Final Result

Before filter:

```text
1000 rows visible
```

After filter:

```text
141 India rows visible
```

Remaining rows become invisible.

---

# Difference Between Column Masking and Row Filtering

| Feature | Column Masking | Row Filtering |
| ------- | -------------- | ------------- |
| Hides   | Column Values  | Entire Rows   |
| Output  | Modified Value | TRUE/FALSE    |
| Example | S***           | Hide USA rows |

---

Example:

This creates layered security.
