# Introduction to Apache Kafka — Lecture Notes (Professor Style)

Welcome back.

In this lecture, we are going to build a strong foundation for understanding Apache Kafka.

We will learn:

* What Apache Kafka is
* Why Kafka is used
* How Kafka works internally
* Kafka producers and consumers
* How Kafka fits into a Data Engineering platform
* Near real-time vs real-time architectures
* Kafka Streams and event-driven systems

So let’s start step by step.

---

# 1. What is Apache Kafka?

Let us begin with the simplest possible definition.

Apache Kafka is an **open-source distributed event streaming platform**.

Now the word *event* may sound complicated initially.

But for practical understanding, whenever you hear the word:

* Event
* Message
* Record

You can think of it simply as:

> “A piece of data.”

So in simple words:

> Kafka is a distributed platform used to collect, store, transport, and process real-time data streams.

---

# 2. What Can We Do Using Kafka?

Kafka mainly solves two major problems.

---

## A. Real-Time Data Collection and Integration

Kafka can collect data continuously from many systems.

Examples:

* Mobile apps
* Websites
* IoT devices
* Sensors
* Databases
* Enterprise applications
* Microservices
* Log files
* Clickstreams

Kafka becomes the central platform where all these systems send data.

So Kafka acts like a **real-time data integration layer**.

---

## B. Real-Time Stream Processing

Once data reaches Kafka:

* Applications can immediately consume it
* Process it
* Analyze it
* Trigger actions
* Generate alerts
* Produce analytics

So Kafka is not only a transport system.

It is also a platform for building streaming applications.

---

# 3. Understanding Kafka Through an Example

Let us understand using a smart meter example.

Imagine:

* Thousands of homes have smart electricity meters
* These meters continuously generate readings
* They send readings every few seconds

Examples of generated data:

* Electricity consumption
* Voltage
* Current load
* Usage patterns

Now think carefully.

Where will all this real-time data go?

This is where Kafka comes into the picture.

---

## Flow of Data

### Step 1 — Devices Generate Data

Smart meters continuously generate readings.

---

### Step 2 — Data Sent to IoT Hub

Sometimes devices cannot directly communicate with Kafka.

So an intermediary system like an IoT Hub is used.

The IoT Hub receives device data.

---

### Step 3 — Kafka Collects the Data

The IoT Hub forwards data into Kafka.

Kafka:

* Receives data
* Stores data
* Organizes data
* Makes it available for processing

---

### Step 4 — Streaming Applications Process Data

A stream-processing application continuously reads data from Kafka.

As soon as new data arrives:

* It gets consumed
* Processed
* Analyzed

All of this can happen within:

* Seconds
* Milliseconds
* Sometimes microseconds

depending on architecture and scale.

---

# 4. Kafka as the Central Hub

This is the most important mental model.

Kafka sits in the middle between:

* Data producers
* Data consumers

It acts as a central broker.

---

# 5. Kafka Architecture

Kafka is designed using the concept of a **messaging queue system**.

There are three main components.

---

# 6. Kafka Core Components

## A. Kafka Broker (Kafka Cluster)

This is the central system.

Kafka brokers are:

* Distributed
* Scalable
* Fault-tolerant

A Kafka cluster may contain:

* 3 servers
* 10 servers
* 100 servers
* Even thousands of servers

depending on scale.

The broker:

* Receives data
* Stores data
* Serves data to consumers

---

## B. Producers

Producers are applications that send data into Kafka.

Examples:

* Mobile applications
* IoT devices
* Banking systems
* Websites
* Microservices

These applications use:

> Kafka Producer APIs

to send messages to Kafka.

---

## C. Consumers

Consumers are applications that read data from Kafka.

They use:

> Kafka Consumer APIs

to receive data.

Once consumers receive data, they can:

* Process it
* Store it
* Analyze it
* Trigger alerts
* Build dashboards

---

# 7. Understanding the Flow

The flow is simple.

---

## Producer Side

Producer applications generate data.

They use Kafka Producer APIs to push data into Kafka.

---

## Kafka Broker

Kafka stores the data safely.

Kafka provides:

* Persistence
* Scalability
* Reliability
* Fault tolerance

---

## Consumer Side

Consumer applications pull data from Kafka using Consumer APIs.

Then they process the data.

---

# 8. Why Kafka is Extremely Powerful

Kafka is highly scalable.

You can have:

* Millions of producers
* Millions of consumers
* Gigabytes of data per second

all working simultaneously.

Kafka scales horizontally.

Meaning:

> Add more servers → handle more traffic.

---

# 9. Kafka in Data Engineering

Now let us connect Kafka with Data Engineering architecture.

A Data Engineering platform generally has three stages:

1. Data Ingestion
2. Data Processing
3. Data Consumption

Kafka mainly fits at the edges.

---

# 10. Kafka at the Data Collection Edge

Imagine many systems generating data:

* Enterprise applications
* Websites
* Mobile apps
* Sensors
* Databases
* Logs
* IoT platforms

Instead of building custom integration for each system:

we use Kafka as the centralized ingestion layer.

All systems send data to Kafka.

This simplifies architecture massively.

---

# 11. Kafka + Lakehouse Architecture

After data reaches Kafka:

you generally have two options.

---

## Option 1 — Directly Create Bronze Tables

A streaming application reads Kafka data directly and creates:

* Bronze tables
* Raw Delta tables

---

## Option 2 — Dump into Landing Zone

Another common pattern:

Kafka → Landing Zone → Bronze Layer

Meaning:

* Consume Kafka data
* Write files into landing zone
* Separate job reads files
* Creates Bronze tables

Both approaches are valid.

---

# 12. Bronze Layer Becomes Source of Truth

Once raw data reaches Bronze tables:

Kafka’s responsibility is mostly complete.

Now the rest of the processing pipeline starts.

Example:

Bronze → Silver → Gold

This is the classic Medallion Architecture.

---

# 13. Kafka at the Consumption Edge

Kafka can also exist at the output side.

After processing:

some consumers may want:

* Tables
* Databases
* APIs

But some applications may want:

> Real-time Kafka streams.

In that case:

your processing application again publishes results back into Kafka.

So Kafka may sit:

* At ingestion
* At consumption
* Or both

---

# 14. Near Real-Time vs Real-Time

This is extremely important.

---

## Near Real-Time Architecture

Typical Data Engineering pipelines:

Kafka → Bronze → Silver → Gold

These pipelines are excellent for:

* Seconds latency
* Minutes latency

This is called:

> Near real-time processing

because data passes through multiple storage layers.

---

## Real-Time Architecture

If requirements demand:

* Millisecond latency
* Microsecond latency

then a different architecture is preferred.

This is where Kafka Streams comes in.

---

# 15. Kafka Streams API

Kafka also provides:

> Kafka Streams Library

This is a Java library used to build real-time stream processing applications.

Using Kafka Streams:

* Microservices consume Kafka data
* Process events instantly
* Produce results immediately

---

# 16. Event-Driven Microservice Architecture

In this architecture:

* Kafka becomes the backbone
* Multiple microservices process events independently

Each microservice:

* Reads events from Kafka
* Performs logic
* Produces new events

This architecture is:

* Distributed
* Highly scalable
* Real-time
* Event-driven

---

# 17. Comparing Both Architectures

## A. Lakehouse / Data Engineering Style

Good for:

* Analytics
* Reporting
* ETL pipelines
* Batch + streaming mix
* Near real-time systems

Latency:

* Seconds to minutes

---

## B. Kafka Streams / Microservices Style

Good for:

* Fraud detection
* Real-time recommendations
* Trading systems
* Instant alerts
* Operational systems

Latency:

* Milliseconds
* Microseconds

---

# 18. Important Mental Model

Remember this carefully.

Kafka is not just a queue.

Kafka is:

* A distributed data integration platform
* A real-time streaming backbone
* A scalable event system
* A persistence layer
* A processing ecosystem

---

# 19. Key Takeaways

## Apache Kafka

* Open-source distributed event streaming platform
* Built for real-time data movement and processing

---

## Kafka Core Components

### Producer

Sends data to Kafka.

### Broker

Stores and distributes data.

### Consumer

Reads data from Kafka.

---

## Kafka Use Cases

* Data integration
* Streaming pipelines
* IoT systems
* Real-time analytics
* Event-driven architectures
* Microservices communication

---

## Kafka in Data Engineering

Kafka generally sits:

* At ingestion edge
* At consumption edge

---

## Two Major Architecture Styles

### Lakehouse Streaming

Near real-time.

### Kafka Streams Microservices

Real-time.

---

# Final Understanding

You should now clearly understand:

* Why Kafka exists
* Why modern systems use Kafka
* How Kafka acts as a real-time backbone
* How producers and consumers interact
* Why Kafka is important in modern Data Engineering

And this foundation is extremely important because later, when we integrate Kafka with Spark Structured Streaming, Delta Lake, Bronze-Silver-Gold architecture, and real-time pipelines, everything will make much more sense.

# Apache Kafka Core Concepts — Professor Style Lecture Notes

Welcome back.

In this lecture, we are going to build a very strong foundation for Apache Kafka by understanding some of the most important core concepts associated with Kafka.

These concepts are extremely important because everything that we do later — producers, consumers, streaming applications, Spark integration, Kafka Streams, real-time pipelines — all of that depends on understanding these fundamentals correctly.

So in this lecture, we will understand these seven core concepts one by one:

1. Kafka Broker or Cluster
2. Producers
3. Consumers
4. Messages
5. Topics
6. Partitions
7. Partition Offsets

Once you understand these concepts properly, Kafka architecture will start making a lot more sense.

So let’s begin.

---

# 1. Kafka Broker or Kafka Cluster

First of all, what is a Kafka Broker?

A Kafka Broker is simply a Kafka server.

But in real-world systems, we usually do not run Kafka on a single machine. Instead, we run Kafka on a cluster of computers.

That is why we often use the term **Kafka Cluster**.

So think of Kafka Cluster as:

* A group of machines
* All running Kafka broker software
* Working together as one distributed system

Once Kafka software is installed and configured, the cluster becomes ready to:

* Receive messages from producers
* Store those messages
* Send those messages to consumers

So Kafka Cluster acts like a central messaging system or central data hub.

---

# 2. Producers

Now let’s move to the producer side.

On the left side of Kafka, we usually have **Producer Applications**.

A producer is nothing but an application that sends data to Kafka.

These applications use something called:

* Kafka Producer API

This API allows applications to connect to Kafka and send messages.

Now one important thing:

Whenever producers send data to Kafka, the data is sent in the form of a **message**.

And every Kafka message generally contains three important fields:

| Field     | Meaning                       |
| --------- | ----------------------------- |
| Key       | Identifier for the message    |
| Value     | Actual data                   |
| Timestamp | Time when message was created |

So every Kafka message is basically:

```text
Key + Value + Timestamp
```

Very important concept.

---

# Understanding Kafka Messages

Let’s understand this with examples.

---

## Example 1 — Reading Data from a File

Suppose we create a producer application that reads data from a text file.

Now imagine the file contains:

```text
Line 1
Line 2
Line 3
```

Our producer application can:

* Read each line
* Convert each line into a Kafka message
* Send it to Kafka

For example:

| Key | Value  | Timestamp    |
| --- | ------ | ------------ |
| 1   | Line 1 | Current Time |
| 2   | Line 2 | Current Time |
| 3   | Line 3 | Current Time |

So each line becomes one Kafka message.

---

## Example 2 — Reading Data from a Database Table

Suppose instead of a file, we are reading data from a database table.

Each database row can become one Kafka message.

For example:

| CustomerID | Name | City     |
| ---------- | ---- | -------- |
| 101        | John | New York |

We may use:

* CustomerID as Key
* Entire row as Value
* Current system time as Timestamp

And then send it to Kafka.

So Kafka producers can send data from:

* Files
* Databases
* Applications
* Mobile devices
* Sensors
* APIs
* Websites
* IoT devices

Practically anything.

---

# Multiple Producers

Another important concept:

Kafka can handle many producers simultaneously.

You may have:

* One producer sending invoice data
* Another producer sending customer data
* Another producer sending sensor data

Or:

* Millions of mobile apps sending the same kind of event

Kafka is designed for massive scalability.

So thousands or even millions of producers can send data in parallel.

---

# 3. Consumers

Now let’s move to the right side of Kafka.

Here we have **Consumer Applications**.

Consumers are applications that:

* Read data from Kafka
* Receive messages
* Process the messages
* Perform some action

Consumers use:

* Kafka Consumer API

to connect to Kafka and receive data.

And remember:

The consumer receives the same structure:

```text
Key + Value + Timestamp
```

because that is how Kafka stores messages internally.

---

# Example of Consumer Processing

Suppose a consumer receives invoice data.

After receiving the messages, the consumer might:

* Process invoices
* Calculate totals
* Save results into a database
* Trigger notifications
* Send emails
* Create reports

Anything is possible.

---

# Multiple Consumers

Just like producers, Kafka also supports multiple consumers.

Example:

| Consumer   | Purpose                 |
| ---------- | ----------------------- |
| Consumer A | Save data into database |
| Consumer B | Send notifications      |
| Consumer C | Run analytics           |
| Consumer D | Fraud detection         |

All can read data from Kafka simultaneously.

That is one of Kafka’s biggest strengths.

---

# Full Kafka Flow

Now let’s combine everything.

Imagine:

* Producer 1 reads from database
* Producer 2 reads from files

Both send messages to Kafka.

Then:

* Consumer 1 processes one type of data
* Consumer 2 processes another type

But now a question comes.

How does Kafka separate different types of data?

How does Consumer A know which messages belong to it?

That brings us to the next concept.

---

# 4. Topics

Kafka introduces the idea of a **Topic**.

A Topic is like a category or channel where messages are stored.

You can think of a Topic like a database table.

For example:

| Topic Name   | Data Type            |
| ------------ | -------------------- |
| invoices     | Invoice data         |
| customers    | Customer data        |
| sensors      | Sensor readings      |
| transactions | Payment transactions |

When producers send data, they specify:

```text
Which topic should receive the message
```

Similarly, consumers also specify:

```text
Which topic should be read
```

So topics help separate datasets.

---

# Topic Example

Suppose:

* Blue messages go to Topic A
* Green messages go to Topic B

Then:

* Blue consumer reads Topic A
* Green consumer reads Topic B

This way consumers only receive relevant data.

Very clean separation.

---

# 5. Topic Partitions

Now comes one of the most important concepts in Kafka.

Kafka Topics are partitioned.

Why?

Because Kafka is designed for huge scale.

Suppose millions of records are arriving every second.

Storing all data on one machine becomes impossible.

So Kafka divides a topic into multiple partitions.

---

# Partition Example

Suppose Topic T1 has 4 partitions:

```text
P0
P1
P2
P3
```

Each partition may sit on a different machine in the cluster.

This allows Kafka to:

* Scale horizontally
* Distribute storage
* Distribute load
* Process data in parallel

This is the foundation of Kafka scalability.

---

# How Kafka Decides Which Partition Gets the Message

Very important.

Kafka uses the **message key**.

The broker:

1. Takes the message key
2. Applies a hash function
3. Generates a partition number

Example:

```text
hash(key) % number_of_partitions
```

Suppose topic has 4 partitions.

Possible partition numbers:

```text
0, 1, 2, 3
```

If hash result is 2:

→ Message goes to Partition P2

---

# Critical Kafka Rule

All messages with the same key always go to the same partition.

This is extremely important.

---

# Store ID Example

Suppose:

* Store ID is the message key
* Messages are invoices/orders

Then:

| Store ID  | Partition |
| --------- | --------- |
| Store 101 | P2        |
| Store 101 | P2        |
| Store 101 | P2        |

All invoices from Store 101 go to the same partition.

Why is this useful?

Because:

* Ordering is preserved
* Aggregation becomes easier
* Processing becomes efficient

Very very important Kafka concept.

---

# What if Key is Null?

Sometimes messages may not have keys.

In that case Kafka may use:

* Round Robin distribution

Example:

```text
Message 1 → P0
Message 2 → P1
Message 3 → P2
Message 4 → P3
```

and repeat.

But in real systems, using meaningful keys is usually preferred.

---

# 6. Partition Offset

Now let’s understand Partition Offset.

Inside each partition, messages are stored sequentially.

Kafka assigns a unique number to every message inside a partition.

That number is called:

```text
Partition Offset
```

---

# Offset Example

Suppose Partition P0 contains:

| Offset | Message   |
| ------ | --------- |
| 0      | Message A |
| 1      | Message B |
| 2      | Message C |
| 3      | Message D |

Offsets always:

* Start from 0
* Increase sequentially

Each partition maintains its own offsets independently.

So:

* P0 may have offset 0–100
* P1 may have offset 0–500
* P2 may have offset 0–1000

All independent.

---

# Why Offsets are Important

Offsets help Kafka consumers track:

* Which messages are already processed
* Which messages are pending

Consumers remember offsets to continue reading from the correct position.

This becomes extremely important for:

* Fault tolerance
* Recovery
* Replay
* Exactly-once processing
* Stream processing

We will study all these later.

---

# Final Summary

Let’s quickly summarize everything.

---

## Kafka Broker / Cluster

* Cluster of machines running Kafka
* Central messaging platform
* Stores and distributes messages

---

## Producer

* Application that sends data to Kafka
* Uses Kafka Producer API

---

## Consumer

* Application that reads data from Kafka
* Uses Kafka Consumer API

---

## Message

Each Kafka message contains:

```text
Key + Value + Timestamp
```

---

## Topic

* Logical category/channel
* Similar to a database table
* Separates datasets

---

## Partition

* Topic is divided into partitions
* Enables scalability and parallelism
* Data distributed using message key

---

## Partition Offset

* Sequential index inside partition
* Starts from 0
* Used for tracking messages

---

# Important Mental Model

A very useful way to remember Kafka is:

```text
Kafka = Distributed Real-Time Messaging Database
```

Where:

* Topics behave like tables
* Messages behave like rows
* Partitions provide scalability
* Offsets provide tracking
* Producers insert data
* Consumers read data

Once this mental model becomes clear, Kafka becomes much easier to learn.
