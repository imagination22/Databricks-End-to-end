Welcome back everyone.

* Source control
* CI/CD pipelines
* Multiple environments
* Automated deployments
* Team collaboration
* Code reviews
* Dev, staging, and production releases

# What are Declarative Automation Bundles?

Earlier, these were called **Asset Bundles**.

Recently, Databricks renamed them to:

# Declarative Automation Bundles

These bundles enable a **Software Engineering Lifecycle** for Data Engineering and AI projects.

---

# Why Were Bundles Introduced?

Before Asset Bundles existed:

* We could store notebooks in GitHub
* We could version-control code files
* But managing jobs, workflows, dashboards, clusters, and deployment configurations was difficult

There was no clean way to package an entire Databricks project together.

Databricks solved this problem using Declarative Automation Bundles.

---

# Main Purpose of Bundles

Bundles help us:

## 1. Treat Databricks Projects as Code

Everything becomes manageable through configuration and version control.

---

## 2. Enable Source Control

Using Git repositories such as:

* GitHub
* Azure DevOps
* Other Git providers

---

## 3. Enable CI/CD Automation

Bundles help automate deployment across environments like:

* Dev
* Staging
* Production

---

# What Can Be Included Inside a Bundle?

A Databricks bundle can contain almost everything related to a project.

Examples include:

## Workspace and Cluster Settings

Infrastructure configurations can be stored.

---

## Notebooks and Python Files

Actual business logic and transformation code.

---

## ML and AI Models

Machine learning artifacts can also become part of the bundle.

---

## Unit Tests and Integration Tests

Testing frameworks and test cases can be included.

---

## Lakeflow Declarative Pipelines

Modern Databricks pipelines are supported.

---

## Jobs and Workflows

Scheduled jobs and orchestration configurations.

---

## Dashboards

Visualization artifacts can also be bundled.

---

# Important Understanding

The instructor explains a very important practical concept.

Suppose we are working on a project that contains:

* Multiple notebooks
* Jobs
* Dashboards
* Pipelines
* Configurations

Instead of managing them separately, we:

# Bundle them together as a single deployable unit

This allows easy deployment across environments.

---

# Real Benefit of Bundles

Without bundles:

* Moving projects between environments is difficult
* Manual setup is required
* Human errors increase

With bundles:

* Everything becomes reproducible
* Deployment becomes automated
* Version control becomes easier
* Collaboration improves

---

# Understanding CI/CD in Real Databricks Projects

Now the lecture moves to the next major topic:

# CI/CD Process

CI/CD stands for:

* Continuous Integration
* Continuous Deployment

The instructor explains how real enterprise projects manage development and deployment.

---

# Step 1 — Create a Git Repository

Every professional project starts with a repository.

Repositories are created in tools such as:

* GitHub
* Azure DevOps
* Other Git platforms

For this demo, GitHub is used.

---

# Main Branch

Whenever a repository is created:

* A default branch is automatically created
* Usually called:

# main

The `main` branch represents:

# Production branch

This is considered the stable final codebase.

---

# Connecting Repository to Databricks

Since the project is related to Databricks:

* The GitHub repository is linked with a Databricks workspace

This allows Databricks to directly interact with repository code.

---

# Development Process Using Branches

The instructor explains a very important industry-standard branching strategy.

---

# Dev Branch

We create another branch from `main` called:

# dev

Purpose:

* Final development branch
* Integration branch for development environment

---

# Important Rule

Developers usually do NOT work directly on the `dev` branch.

Instead:

They create separate feature branches.

---

# Feature Branch Workflow

Suppose a developer wants to build a new feature.

They create:

# feature branch

from the `dev` branch.

Example:

* feature1
* feature_customer_pipeline
* feature_dashboard_upgrade

---

# Development Happens in Feature Branches

Inside feature branches developers:

* Modify notebooks
* Add jobs
* Create workflows
* Update dashboards
* Add tests

Once development is complete:

* Testing is performed
* Pull Request is created
* Code is reviewed
* Changes are merged into `dev`

---

# Parallel Development

The instructor explains that multiple developers can work simultaneously.

Example:

| Developer   | Branch   |
| ----------- | -------- |
| Developer 1 | feature1 |
| Developer 2 | feature2 |
| Developer 3 | feature3 |

This enables:

* Parallel development
* Faster delivery
* Independent feature implementation

---

# Deployment to Dev Environment

After merging into `dev`:

The project bundle is deployed into:

# Dev Workspace

This helps test the deployed version before production release.

---

# Deployment Options

Two approaches exist:

## Manual Deployment

Developer manually clicks deploy.

---

## Automated Deployment

Using:

# GitHub Actions

---

# Dev Deployment Practices

Different companies follow different practices.

Some projects:

* Use manual deployment in Dev

Others:

* Fully automate Dev deployment

Both are valid depending on project requirements.

---

# Testing in Dev Environment

After deployment:

* Jobs are executed
* Pipelines are validated
* Testing is performed

Once everything works correctly:

The code is considered production-ready.

---

# Production Release Process

Now a Pull Request is created:

# dev → main

This is one of the most important stages.

---

# Code Review Process

Before merging into production:

* Senior engineers
* Architects
* Technical reviewers

review the changes.

Only after approval:

* Changes are merged into `main`

This ensures:

* Code quality
* Stability
* Governance
* Best practices

---

# Production Deployment

Once code reaches the `main` branch:

The bundle is deployed into:

# Production Workspace

---

# First-Time Deployment Practice

The instructor explains a real-world best practice.

For first deployment:

# Manual deployment is preferred

Reason:

* Validate process
* Understand errors
* Ensure end-to-end setup works properly

---

# Later Releases

Once the process becomes stable:

# GitHub Actions automate deployment

Now deployment happens automatically whenever:

* Pull Request is merged
* Code enters main branch

No manual deployment required.

---

# Development Lifecycle Repeats

Software development is continuous.

After one release:

* New enhancements come
* New features are requested
* Bugs are fixed

So again:

* New feature branch created
* Changes developed
* PR raised
* Review completed
* Dev deployment done
* Testing completed
* Production deployment performed

This cycle continues repeatedly.

---

# Role of GitHub Actions

GitHub Actions automate deployment workflows.

Once configured:

* Pull Request merged
* Deployment triggered automatically
* Bundles deployed without manual action

This provides:

* Faster delivery
* Reduced manual effort
* Standardized deployment
* Reliable CI/CD pipelines

---

Now the instructor starts practical implementation.

---

# Step 1 — Create GitHub Repository

Inside GitHub:

* Click “New Repository”
* Provide repository name

Example used:

# databricks-asset-bundle-bakehouse-analytics

---

# Visibility Setting

The instructor explains an important best practice.

Although GitHub defaults to:

# Public Repository

In real projects we should use:

# Private Repository

Reason:

* Protect source code
* Maintain security
* Restrict external visibility

---

# Repository URL

After repository creation:

GitHub provides a repository URL.

This URL is important because:

# Databricks uses this URL to connect with the repository

---

# Connecting GitHub to Databricks Workspace

Inside Databricks:

1. Open Workspace
2. Go to Settings
3. Open Linked Accounts
4. Configure Git Integration

---

# Git Provider Options

Databricks supports multiple providers:

* GitHub
* Azure DevOps
* Others

For this demo:

# GitHub is selected

---

# Authentication

The instructor authenticates GitHub account.

After successful authentication:

Databricks workspace becomes linked with GitHub.

This means:

* Databricks can pull code
* Databricks can push changes
* Repositories can be synchronized

---

# Creating Git Folders in Databricks

Databricks allows two types of folders:

## Normal Workspace Folder

Regular Databricks folder.

---

## Git Folder

Connected directly to a Git repository.

For CI/CD projects:

# Git Folders are used

---

# Creating Production Folder

The instructor first creates a Git folder for:

# Production Environment

Initially it links to:

# main branch

---

# Installing Databricks Git App

Databricks requires repository permissions.

An app/extension must be installed inside GitHub repository.

Permissions can be granted:

* For all repositories
* Or selected repositories

Best practice:

# Grant only required repositories

---

# Dev Folder Creation

Next:

Another Git folder is created for:

# Dev Environment

Initially it also points to `main` because no additional branches exist yet.

---

# Problem Encountered

While creating the `dev` branch:

An error occurs:

# Cannot create branch from empty repository

Reason:

No initial commit exists.

---

# Solution

The instructor fixes it by:

1. Creating a README file
2. Making first commit
3. Pushing changes

After first commit:

Branch creation becomes possible.

---

# Understanding Commit

A commit represents:

# Saved version of changes

It captures modifications relative to previous version.

---

# Pulling Changes

Inside Databricks Git folder:

A `pull` operation synchronizes latest repository changes into workspace.

This fetches:

* README
* Latest commits
* Updated branch contents

---

# Creating Dev Branch Successfully

Now the `dev` branch is created successfully.

The workspace is refreshed and the branch becomes visible.

---

# Creating Feature Branch

Following proper CI/CD practice:

A new feature branch is created from:

# dev branch

Example branch name:

# feature1

This becomes the actual development branch.

---

# Final Understanding

At this stage the project now contains:

| Branch   | Purpose                 |
| -------- | ----------------------- |
| main     | Production              |
| dev      | Development Integration |
| feature1 | Feature Development     |

This is the standard professional workflow followed in enterprise projects.
