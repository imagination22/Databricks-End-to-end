Hi everyone 👋

The goal of this session is not just to memorize diagrams, but to understand:

✅ How Databricks works internally
✅ How accounts, workspaces, and metastore are connected

# 🏢 1. Databricks Starts with an Account

In Databricks, everything begins with an **Account Level**.

Every organization first creates a:

# 👉 Databricks Account

Inside that account, we manage:

* 🖥️ Workspaces
* 📚 Unity Catalog Metastores
* 👤 Users & Groups
* 🔐 Security
* 💳 Billing & Usage
* 🌍 Multi-region management

So the **Account** acts like the **central management layer**.

# 🧠 Important Mental Model

Think like this:

| Component | Purpose                                 |
| --------- | --------------------------------------- |
| Account   | Central control room 🏢                 |
| Workspace | Place where developers work 💻          |
| Metastore | Central governance & metadata system 📚 |

---

# 🏗️ Account-Level Responsibilities

At the account level, Databricks manages many important things.

---

## 👥 1. Identity and Access Management

At account level, we manage:

* Users
* Groups
* Service Principals
* Permissions

This is where authentication and authorization are controlled 🔐

Example:

* Creating new developer accounts
* Adding users to groups
* Managing admin roles

---

## 🌍 2. Workspace Management

An organization can have:

* Multiple workspaces
* Across multiple regions

Example:

| Region | Workspace            |
| ------ | -------------------- |
| India  | Dev Workspace        |
| US     | Production Workspace |
| Europe | Analytics Workspace  |

All of them are controlled from a **single account** 🌎

At account level we can:

✅ Create workspace
✅ Update workspace
✅ Delete workspace

---

## 📚 3. Unity Catalog Metastore Management

Databricks uses **Unity Catalog** for centralized governance.

The metastore manages:

* Tables
* Schemas
* Databases
* ML Models
* Volumes
* Permissions

Think of Unity Catalog as the **central library manager** 📖

---

# 🧩 Important Rule About Metastore

## 🛑 One Region = One Metastore

In one region:

✅ Only one Unity Catalog Metastore can exist

But:

✅ Multiple workspaces can connect to the same metastore

---

# 🎯 Example Architecture

Imagine:

## Region: India 🇮🇳

You create:

* One Unity Catalog Metastore
* Three Workspaces

Example:

| Workspace | Purpose     |
| --------- | ----------- |
| Dev       | Development |
| QA        | Testing     |
| Prod      | Production  |

All three can use the same metastore 📚

This ensures:

* Central governance
* Consistent permissions
* Shared metadata

---

A Workspace is basically the environment where developers and data engineers work daily.

This is the actual portal/interface you interact with.

Inside workspace we do:

✅ Write notebooks
✅ Run Spark code
✅ Create jobs
✅ Perform ML training
✅ Interactive analytics
✅ Data ingestion
✅ Schedule pipelines

# 🏛️ Databricks Architecture Overview
![image_1778651323599.png](./image_1778651323599.png "image_1778651323599.png")

Now let us go one level deeper.

Databricks architecture is divided into two major parts:

# 1️⃣ Control Plane

# 2️⃣ Compute Plane

This is one of the MOST IMPORTANT interview concepts 🚨

---

# 🎛️ 1. Control Plane

The Control Plane is managed entirely by Databricks.

It contains the intelligence and orchestration layer.

---

# 📦 Components Inside Control Plane

## 🌐 Web Application

The Databricks UI/portal that we use daily.

Example:

* Workspace UI
* Notebook interface
* Job UI
* SQL editor

Everything belongs to the control plane.

---

## ⚙️ Compute Orchestration

When we create clusters:

* Start cluster
* Stop cluster
* Scale cluster
* Monitor cluster

All this orchestration happens from the control plane.

---

## 📚 Unity Catalog

Unity Catalog is also managed in the control plane.

It stores:

* Metadata
* Governance rules
* Permissions

---

## 📒 Notebooks & Jobs Metadata

The following are also part of control plane:

* Notebook definitions
* Job definitions
* Query metadata

---

# ☁️ 2. Compute Plane

The Compute Plane is where actual computation happens ⚡

This includes:

* Spark execution
* Cluster VMs
* Data processing

---

# 🖥️ Job Cluster / All-Purpose Cluster Architecture

If we use:

✅ Job Cluster
✅ All-Purpose Cluster

Then compute runs in our cloud environment.

Example:

* Azure VMs
* AWS EC2
* GCP Compute Engine

---

# 🔄 How It Works

## Step-by-Step Flow

### Step 1

User writes notebook in workspace ✍️

### Step 2

Control plane receives request 🎛️

### Step 3

Control plane creates/manages cluster ⚙️

### Step 4

Cloud provider creates VMs ☁️

### Step 5

Spark job executes on those VMs ⚡

---

# 📦 Workspace Storage Account

The actual storage generally belongs to customer cloud environment.

Example:

* Azure Data Lake Storage
* AWS S3
* Google Cloud Storage

This storage is NOT inside Databricks control plane.

It belongs to YOUR cloud account ☁️

---

# 🚀 Serverless Compute Architecture
![image_1778651353406.png](./image_1778651353406.png "image_1778651353406.png")
Now comes modern architecture:

# 👉 Serverless Compute

In serverless:

Databricks itself manages all infrastructure internally.

You do NOT see:

❌ Virtual Machines
❌ Cluster infrastructure
❌ Scaling management

Everything becomes hidden from users 🔥

---

# 🧠 Key Idea

With serverless:

✅ Databricks controls compute
✅ Databricks manages scaling
✅ Databricks optimizes infrastructure

You simply run workloads.

---

# 📌 Difference Between Classic vs Serverless

| Feature                | Classic Workspace | Serverless Workspace |
| ---------------------- | ----------------- | -------------------- |
| Cluster Visibility     | Visible           | Hidden               |
| VM Management          | Customer Cloud    | Databricks           |
| All-Purpose Clusters   | Supported         | Not Preferred        |
| Job Clusters           | Supported         | Mostly Serverless    |
| Infrastructure Control | More              | Less                 |
| Simplicity             | Medium            | High                 |
| Flexibility            | High              | Medium               |

---

# 🆕 Complete Serverless Workspace

Recently Databricks introduced:

# 🌟 Complete Serverless Workspace

In this model:

✅ Only serverless compute is recommended
✅ Traditional clusters are removed/minimized
✅ Databricks manages almost everything

---

# 📦 Storage in Serverless Workspace

Databricks also provides:

# 👉 Default Storage

Internally it may still use Azure/AWS/GCP storage, but users do not directly manage it.

This simplifies setup significantly.

---

# 🤔 Why Databricks is Moving Towards Serverless?

Because Databricks wants:

✅ Better user experience
✅ Easier infrastructure management
✅ Faster startup times
✅ Full platform optimization
✅ Unified billing and control

But pricing will become an important deciding factor 💰

---

# 🧠 Final Big Picture

# 🏢 Account Level

Controls everything globally.

Contains:

* Workspaces
* Metastore
* Security
* Billing

↓

# 💻 Workspace

Developer collaboration environment.

Contains:

* Notebooks
* Jobs
* Compute access

↓

# 🎛️ Control Plane

Managed by Databricks.

Contains:

* UI
* Orchestration
* Metadata
* Unity Catalog

↓

# ⚡ Compute Plane

Where actual Spark execution happens.

Can be:

* Customer cloud VMs
* Databricks serverless infrastructure

## 1. What is a Databricks Notebook?

A Databricks notebook is an interactive environment for:

- Writing and running **code** (Python, SQL, Scala, R)
- Combining **code + markdown** documentation
- **Experimenting** with data
- **Visualizing** results
- Collaborating with others

Each notebook is made up of **cells**:
- **Code cells** – run code (Python / SQL / etc.)
- **Markdown cells** – formatted text, images, headings

You can:
- Add a cell – using the `+` button or keyboard shortcuts
- Move cells – drag the left side handle
- Run a single cell – `Shift + Enter`
- Run all cells – from the `Run all` or `Run selected` menu

## 2. Using Multiple Languages with Magic Commands

A Databricks notebook has a **default language** (e.g., Python),  
but you can run other languages in any cell using **magic commands**:

- `%python` – Python
- `%sql` – SQL
- `%scala` – Scala
- `%r` – R
- `%md` – Markdown inside a code cell
- `%sh` – Shell commands
- `%fs` – Filesystem commands
- `%pip` – Install Python packages

You put the magic command at the **top of the cell**, as the first line.

## 4. Accessing Filesystem with `%fs`

Databricks provides a special filesystem abstraction called **DBFS (Databricks File System)**.

Common `%fs` commands:

- `%fs ls` – list files/directories  
- `%fs mkdirs /mnt/mydir` – create directories  
- `%fs rm -r /mnt/mydir` – remove directory  
- `%fs head /path/file` – preview file

You can also use `dbutils.fs` in Python.


# COMMAND ----------

# DBTITLE 1,FS Magic Command
%fs ls /Volumes/my_catalog_2/my_schema/my_volume/my_folder/

# COMMAND ----------

# DBTITLE 1,FS using DB Utils
# 6. Using dbutils.fs in Python

# List files in root
dbutils.fs.ls("/Volumes/my_catalog_2/my_schema/my_volume/my_folder/")

## 11. Widgets (Interactive Notebook Parameters)

Widgets allow you to **parameterize** your notebook.

Types of widgets:
- `text` – free text
- `dropdown` – one value from a list
- `combobox` – Combination of text and dropdown
- `multiselect` – multiple values

Core APIs:
- `dbutils.widgets.text(name, defaultValue, label)`
- `dbutils.widgets.dropdown(name, defaultValue, choices, label)`
- `dbutils.widgets.get(name)` – to read the value
- `dbutils.widgets.remove(name)` / `removeAll()`

Widgets are great for:
- Running the same notebook with different inputs
- Using with **Jobs** for scheduled runs


# COMMAND ----------

# DBTITLE 1,Widgets
# 12. Creating widgets

dbutils.widgets.removeAll()

dbutils.widgets.text("p_name", "Alice", "Employee name")
dbutils.widgets.dropdown("p_min_salary", "0", ["0", "500", "1000"], "Min Salary")

print("Widgets created. Change the values in the top widget panel and re-run the next cell.")

# COMMAND ----------

# 13. Reading widget values and applying filters

name = dbutils.widgets.get("p_name")
salary = float(dbutils.widgets.get("p_min_salary"))

print(f"name: {name}, salary: {salary}")

- Syntax: `%run /Path/To/OtherNotebook`
- All functions, variables, and classes from that notebook become available

Useful for:
- Creating **utility** notebooks (e.g., common functions)
- Keeping code modular and organized

## 13. Revision History & Checkpoints (Concept)

Databricks notebooks support:

- **Revision history**:
  - View, compare, and restore previous versions

- **Autosave**:
  - Notebooks are saved automatically as you edit

- **Export & Import**:
  - Export as `.dbc`, `.html`, or `.ipynb` (depending on workspace)
  - Import notebooks from files or from Git

You can use this to roll back experiments, review changes, and share work.
